```python
import requests
from lxml import html
import datetime
import pandas
```

Pars Mail.ru

```python
u_a = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:69.0) Gecko/20100101 Firefox/69.0',
           'Accept':'*/*'}
source = 'https://mail.ru'

req = requests.get(source,headers = u_a).text
page = html.fromstring(req)
'''
if page.xpath('//div[@class="news-item__inner"]/a[2]'):
    link=page.xpath('//div[@class="news-item__inner"]/a[2]/@href')
    news_title = page.xpath('//div[@class="news-item__inner"]/a/@href/text()')
else:'''
news_data = []
for news in page.xpath('//div[@class="news-item__inner"]'):
    news_data.append({'news_link': news.xpath('a/@href'),
                      'news_title': news.xpath('a/text()'.replace('\xa0', ' ')),         
                      'news_date': datetime.date.today().strftime('%d/%m/%Y'),
                      'news_source': source}) 
```

pars lenta.ru

```python
source_lenta = 'https://lenta.ru'
req_l = requests.get(sourse_lenta, headers=u_a).text

lenta_page = html.fromstring(req_l)

for news in lenta_page.xpath('//div[contains(@class, "span4")]/div[contains(@class,"item")]'):
    news_data.append({'news_link': news.xpath('a/@href|h2/a/@href'),
                      'news_title': news.xpath('a/text()|h2/a/text()'.replace('\xa0', ' ')),         
                      'news_date': news.xpath('a/time/@datetime|h2/a/time/@datetime'),
                      'news_source': source_lenta})
```
