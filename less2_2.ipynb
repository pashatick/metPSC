```python
import requests as req
import pprint
from bs4 import BeautifulSoup as bs
import pandas as pd
import getpass
import pprint
import numpy as np
```

```python
main_link = 'https://roscontrol.com'
link = '/category/produkti/'

name_ = []
cat_name1_ = []
cat_name2_ = []
cat_name3_ = []
security_ = []
quality_ = []
ratio_ = []
url_ = []

def pars (link):
    cat = req.get(main_link+link)
    cat.encoding='utf-8'
    cat = bs(cat.text, 'html.parser')
    return cat

cat_list = pars(link).findAll('div',{'class':'grid-padding grid-column-3 grid-column-large-6 grid-flex-mobile grid-column-middle-6 grid-column-small-12 grid-left'})

for category in cat_list:
    
    cat_link = category.find('a', {'class':'catalog__category-item util-hover-shadow'})['href']
    sub_cat_lis = pars(cat_link).findAll('div',{'class':'grid-padding grid-column-3 grid-column-large-6 grid-flex-mobile grid-column-middle-6 grid-column-small-12 grid-left'})
    
    for sub_category in sub_cat_lis:
        
        sub_link = sub_category.find('a', {'class':'catalog__category-item util-hover-shadow'})['href']
        prod_list = pars(sub_link).findAll('div', {'class':'wrap-product-catalog__item grid-padding grid-column-4 grid-column-large-6 grid-column-middle-12 grid-column-small-12 grid-left js-product__item'})
        
        for prod in prod_list:

            prod_link = prod.find('a', {'class':'block-product-catalog__item js-activate-rate util-hover-shadow clear'})['href']
            pd = pars(prod_link)
            
            name_.append(pd.find('h1',{'class':'main-title testlab-caption-products util-inline-block'}).getText())
            cret = pd.findAll('div', {'class':'rate-item group'})
            url_.append(main_link+prod_link)
            
            if len(pd.find('div',{'id':'product__single-rev-rate'}))>5:
                ratio_.append(pd.find('div',{'id':'product__single-rev-total'}).find('div', {'class':['total green', 'total orange']}).getText())
                quality_.append(cret[-1].find('div',{'class':'rate-item__value'}).getText().replace(' ',''))
                security_.append(cret[0].find('div',{'class':'rate-item__value'}).getText().replace(' ',''))
            else:
                ratio_.append('Black list')#(pd.find('div',{'class':'product__single-subtitle'}).find('span',{'class':'not-show show-with-1024'}).getText())
                quality_.append('Нет данных')
                security_.append('Нет данных')

            
            cat_name = pd.findAll('li',{'itemprop':'itemListElement'})
            cat_name1_.append(cat_name[0].find('span',{'itemprop':'name'}).getText()) 
            cat_name2_.append(cat_name[1].find('span',{'itemprop':'name'}).getText())
            cat_name3_.append(cat_name[2].find('span',{'itemprop':'name'}).getText())
        #break
    break
```

```python
import pandas
data={'Наименование продукта':name_,
      'Категория продукта':cat_name1_,                
      'Подкатегория продукта':cat_name2_,
      'Тип продукта':cat_name3_,
      'Параметр "безопасность"':security_,
      'Параметр "качество"':quality_,
      'Общий балл':ratio_,
      'Сайт':url_}
prod_data = pandas.DataFrame(data)
prod_data
```
